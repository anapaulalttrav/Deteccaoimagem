```python
import cv2
import matplotlib.pyplot as plt


# Carrega o classificador Haar para
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
# Carrega a imagem
lu = cv2.imread('/content/fotolu.JPG')
gray = cv2.cvtColor(lu, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(lu)
```

```python
plt.imshow(gray)
```

```python
# Detecta faces
faces = face_cascade.detectMultiScale(gray, 1.1, 5)
```

```python
# Desenha retângulos ao redor das faces detectadas
for (x, y, w, h) in faces:
    cv2.rectangle(lu, (x, y), (x+w, y+h), (255, 0, 0), 2)
```

```python
plt.imshow(cv2.cvtColor(lu, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib
```
